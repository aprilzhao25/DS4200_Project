```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
df = pd.read_csv('healthcare_dataset_cleaned.csv')
df.head()
```

```python
avg_eoc = df.groupby('Gender')['Duration of Stay'].mean()
```

```python
plt.bar(avg_eoc.index, avg_eoc.values, color=['blue', 'brown'])
plt.xlabel('Gender')
plt.ylabel('Average EOC Duration')
plt.title('Average Stay of Duration by Gender')
```

```python
for val in df['Insurance Provider'].unique():
    dfs = df[df['Insurance Provider'] == val]
    plt.hist(dfs['Billing Amount'], bins = 5, alpha = 0.5)
```

```python
ins_grouped2 = df.groupby('Insurance Provider')['Billing Amount'].mean()
ins_grouped2.plot(kind = 'bar')
print(ins_grouped2)
```

```python
ins_grouped2 = df.groupby('Insurance Provider')['Billing Amount'].sum()
ins_grouped2.plot(kind = 'bar')
print(ins_grouped2)
```
